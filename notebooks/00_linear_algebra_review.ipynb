# 00 — Linear Algebra Review

This notebook reviews the linear algebra fundamentals that underpin deep learning:
- Vectors, matrices, and tensors
- Matrix multiplication
- Eigenvalues and eigenvectors
- Singular value decomposition (SVD)

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Vectors and Matrices

```python
# Column vector
v = np.array([1, 2, 3], dtype=float)
print('v =', v)
print('shape:', v.shape)

# Matrix
A = np.array([[1, 2], [3, 4], [5, 6]], dtype=float)
print('\nA =\n', A)
print('shape:', A.shape)
```

## 2. Matrix Multiplication

```python
B = np.array([[1, 0], [0, 1]], dtype=float)  # Identity
C = A @ B
print('A @ I =\n', C)
```

## 3. Eigenvalues and Eigenvectors

```python
M = np.array([[4, 1], [2, 3]], dtype=float)
eigenvalues, eigenvectors = np.linalg.eig(M)
print('Eigenvalues:', eigenvalues)
print('Eigenvectors:\n', eigenvectors)
```

## 4. Singular Value Decomposition

```python
U, S, Vt = np.linalg.svd(A, full_matrices=False)
print('U:\n', U)
print('S:', S)
print('Vt:\n', Vt)
print('Reconstruction error:', np.linalg.norm(A - U @ np.diag(S) @ Vt))
```
