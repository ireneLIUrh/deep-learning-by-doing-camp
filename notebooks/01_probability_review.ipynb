# 01 — Probability Review

This notebook reviews probability concepts essential for deep learning:
- Random variables and distributions
- Expectation and variance
- Bayes' theorem
- Common distributions (Gaussian, Bernoulli, Categorical)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

## 1. Gaussian Distribution

```python
mu, sigma = 0.0, 1.0
x = np.linspace(-4, 4, 200)
pdf = stats.norm.pdf(x, mu, sigma)

plt.figure(figsize=(6, 3))
plt.plot(x, pdf)
plt.title('Standard Normal PDF')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.tight_layout()
plt.show()
```

## 2. Expectation and Variance

```python
rng = np.random.default_rng(42)
samples = rng.normal(mu, sigma, size=10000)
print(f'Empirical mean:     {samples.mean():.4f}  (expected {mu})')
print(f'Empirical variance: {samples.var():.4f}  (expected {sigma**2})')
```

## 3. Bayes' Theorem

```python
# Toy example: disease test
p_disease = 0.01        # prior
p_pos_given_disease = 0.95  # sensitivity
p_pos_given_no_disease = 0.05  # false positive rate

p_pos = p_pos_given_disease * p_disease + p_pos_given_no_disease * (1 - p_disease)
p_disease_given_pos = (p_pos_given_disease * p_disease) / p_pos
print(f'P(disease | positive test) = {p_disease_given_pos:.4f}')
```

## 4. Categorical Distribution & Cross-Entropy

```python
# Softmax probabilities
logits = np.array([2.0, 1.0, 0.1])
probs = np.exp(logits) / np.exp(logits).sum()
print('Probabilities:', probs)

# Cross-entropy against true label 0
true_label = 0
cross_entropy = -np.log(probs[true_label])
print(f'Cross-entropy loss: {cross_entropy:.4f}')
```
