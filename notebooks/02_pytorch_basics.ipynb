# 02 — PyTorch Basics

This notebook introduces core PyTorch concepts:
- Tensors and operations
- Autograd and backpropagation
- Building a simple neural network with `nn.Module`
- A minimal training loop

```python
import torch
import torch.nn as nn
import torch.optim as optim
print('PyTorch version:', torch.__version__)
```

## 1. Tensors

```python
x = torch.tensor([[1.0, 2.0], [3.0, 4.0]])
print('x:\n', x)
print('shape:', x.shape)
print('dtype:', x.dtype)

# Element-wise operations
print('x * 2:\n', x * 2)
print('x @ x.T:\n', x @ x.T)
```

## 2. Autograd

```python
w = torch.tensor(2.0, requires_grad=True)
b = torch.tensor(1.0, requires_grad=True)

x_in = torch.tensor(3.0)
y_pred = w * x_in + b
y_true = torch.tensor(10.0)

loss = (y_pred - y_true) ** 2
loss.backward()

print(f'loss: {loss.item():.4f}')
print(f'dL/dw: {w.grad.item():.4f}')
print(f'dL/db: {b.grad.item():.4f}')
```

## 3. Simple `nn.Module`

```python
class TinyMLP(nn.Module):
    def __init__(self, in_features: int, hidden: int, out_features: int):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(in_features, hidden),
            nn.ReLU(),
            nn.Linear(hidden, out_features),
        )

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.net(x)

model = TinyMLP(4, 8, 2)
print(model)
sample = torch.randn(5, 4)
out = model(sample)
print('output shape:', out.shape)
```

## 4. Minimal Training Loop

```python
torch.manual_seed(0)
model = TinyMLP(4, 8, 2)
optimizer = optim.Adam(model.parameters(), lr=1e-3)
criterion = nn.CrossEntropyLoss()

X = torch.randn(100, 4)
y = torch.randint(0, 2, (100,))

for epoch in range(10):
    optimizer.zero_grad()
    logits = model(X)
    loss = criterion(logits, y)
    loss.backward()
    optimizer.step()
    if (epoch + 1) % 2 == 0:
        print(f'Epoch {epoch+1:3d} | loss: {loss.item():.4f}')
```
